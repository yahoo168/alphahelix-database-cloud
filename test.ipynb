```python
# 將requirements.txt轉為setup.py的install_requires
def generate_install_requires():
    with open('requirements.txt', 'r') as f:
        lines = f.readlines()
    
    install_requires = []
    for line in lines:
        # 去除注释和空行
        line = line.strip()
        if line and not line.startswith('#'):
            install_requires.append(f'"{line}"')

    # 打印格式化后的 install_requires 列表
    install_requires_str = ",\n".join(install_requires)
    print(f"install_requires=[\n{install_requires_str}\n]")
    
generate_install_requires()
```
